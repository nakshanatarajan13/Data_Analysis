```python
import pandas as pd
data=pd.read_csv("movies.csv")
data.head()
```

```python
data.info()
```

#### What is the average imdb score for movies? 

```python
grp1=data.groupby(["type"])["imdb_score"].mean()
print(grp1)
```

#### How many null values are there in the ‘Age_certification’ column.

```python
nullvalues=data["age_certification"].isna().sum()
print("The null values in the column the age_certification is:",nullvalues)
```

#### How many numeric valued columns are present?

```python
import numpy as np
numeric=data.select_dtypes(include=np.number)
print("The number of numeric column is:",len(numeric.columns))
```

#### Write the code to display the movies released after the year 2000. How many movies were released after the year 2000?

```python
column=data["release_year"]
count=column[column>2000].count()
print("The movies after 2000 are:",count)
```

#### counting the null values:

```python
print(data.isna().sum())
```

#### After dropping all the null values in the dataset, how many rows remain?

```python
print("The number of rows before removing null values:",len(data.index))
```

```python
newdf=data.dropna()
print("The number of rows after removing null values:",len(newdf.index))
```

#### Out of these rows, how many movies released after 2020?

```python
column=newdf["release_year"]
count=column[column>2020].count()
print("The movies after 2020 are:",count)
```

#### What is the average tmdb_popularity for movies containing ‘Comedy’ as a genre?

```python
grp2=newdf.groupby(["genres"])["tmdb_popularity"].mean()
print(grp2)
```

```python
newdf[(newdf["genres"].str.contains("comedy"))].describe().T.loc["tmdb_popularity"].loc["mean"]
```
